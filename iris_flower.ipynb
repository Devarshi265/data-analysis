```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv'
df = pd.read_csv(url)

print('few rows of dataser')
print(df.head())

print('\n stats of dataset:')
print(df.describe())
```

```python
print('\n info about dataset')
print(df.info())
```

```python
plt.figure(figsize=(10,6))
sns.pairplot(df,hue='species',palette='Set1')
plt.title('pair plot')
plt.show()


plt.figure(figsize=(10,6))
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df)
plt.title('boxplot')
plt.show()

corr_matrix = df.drop('species',axis=1).corr()
print('\n correlation matrix')
print(corr_matrix)

species_mean = df.groupby('species').mean()
print('\n mean values of features by species')
print(species_mean)

plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot = True, cmap='coolwarm')
plt.title('corr heatmap')
plt.show()

print('\n summary of findings:')
print("1, dataset contains 3 species of iris flowers: setosa,varsicolor and virginica.")

```
